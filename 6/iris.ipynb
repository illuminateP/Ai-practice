```python
# /content/drive/MyDrive/Colab Notebooks/classifier/iris.csv
filename = '/content/drive/MyDrive/Colab Notebooks/classifier/iris.csv'
data = pd.read_csv(filename)
print(data.info())
print(data.head())
```

```python
# 중복을 허용하지 않고 있는 데이터의 모든 종류를 출력
print("품종 종류:", data["variety"].unique(), sep="\n")

# 라벨을 정수로 인덱싱
data['variety'] = data['variety'].replace(['Virginica','Setosa','Versicolor'],[0,1,2])
print("품종 종류:", data["variety"].unique(), sep="\n")
```

```python
data_X = data[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']].values
data_y = data['variety'].values
print(data_X[:5])
print(data_y[:5])
```

```python
(train_X, test_X, train_Y, test_Y) = train_test_split(data_X, data_y, train_size=0.8)
print(train_X.shape)
print(train_Y.shape)
print(test_X.shape)
print(test_Y.shape)
```

```python
from keras.utils import to_categorical     # 훈련 데이터와 테스트 데이터에 대해서 원-핫 인코딩
train_Y = to_categorical(train_Y)
test_Y = to_categorical(test_Y)

print(train_Y[:5])
print(test_Y[:5])
```

```python
model= Sequential()
model.add(Dense(3, input_dim=4, activation='softmax'))
model.summary()
```

```python
model.compile(optimizer = optimizers.SGD(0.01), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(train_X, train_Y, batch_size = 1, epochs = 200, validation_data=(test_X, test_Y))
```

```python
model.evaluate(test_X, test_Y)
```

```python
predit = model.predict(test_X)
#for i in range(0,len(predit)) :
#    print("예측값 : ", predit[i], "\t실제값 : ", test_Y[i])
print(predit[0:5])
print(test_Y[0:5])
```

```python
import matplotlib.pyplot as plt
epochs = range(1, len(history.history['accuracy']) + 1)
plt.plot(epochs, history.history['accuracy'])
plt.plot(epochs, history.history['val_accuracy'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
print("\n 테스트 정확도: %.4f" % (model.evaluate(test_X, test_Y)[1]))
```
