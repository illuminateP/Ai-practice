```python
import numpy as np                                        # Numpy를 임포트
import matplotlib.pyplot as plt
```

```python
z = np.arange(0.00001, 1, 0.01)         # x는 예측값     
cost = -np.log(z)                       # y가 1일 때 cost
```

```python
plt.plot(z, cost, 'b', label = 'real value y =1')       
plt.xlabel('x')                     # x축 라벨 
plt.ylabel('cost')                  # y축 라벨
plt.legend(loc='upper left')        # 범례
plt.show()   
```

```python
z = np.arange(0.00001, 1, 0.01)           # x는 예측값
cost = -(np.log(1-z))                     # y가 0일 때 cost
```

```python
# y 가 0일 때 파란선
# x 가 0에 가까울 수록 오차가 적다
plt.plot(z, cost, 'r', label = 'real value y = 0')       
plt.xlabel('x')                     # x축 라벨 
plt.ylabel('cost')                  # y축 라벨
plt.legend(loc='upper left')        # 범례
plt.show()   
```

```python
z = np.arange(0, 1, 0.01)           
cost1 = -(np.log(z))      #  실제값이 1일 때 오차
cost0 = -(np.log(1-z))    #  실제값이 0일 때 오차
```

```python
plt.plot(z, cost1, 'b', label = 'real value y =1')       
plt.plot(z, cost0, 'r', label = 'real value y = 0') 
plt.xlabel('predit valuie')                     # x축 라벨 
plt.ylabel('cost')                     # y축 라벨
plt.legend(loc='upper left')                 # 범례
plt.title('cost function for logistic')      # 제목
plt.show()                                   # 그래프 출력
```
