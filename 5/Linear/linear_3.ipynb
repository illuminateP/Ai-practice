```python
from keras.models import Sequential            # 케라스의 Sequential()을 임포트
from keras.layers import Dense                 # 케라스의 Dense()를 임포트
from keras import optimizers                   # 케라스의 옵티마이저(학습 방법)를 임포트
import numpy as np                             # Numpy를 임포트
```

```python
# 입력 값 / 실제 값
x_data = [[73., 80., 75.], [93., 88., 93.], [89., 91., 90.], [96., 98., 100.], [73., 66., 70.]]
y_data = [[152.], [185.], [180.], [196.], [142.]]
```

```python
model = Sequential()
model.add(Dense(1, input_dim=3, activation='linear'))
model.summary()
```

```python
model.compile(optimizer = optimizers.Adam(lr = 0.00001), loss='mse')

hist = model.fit(x_data, y_data, batch_size =1, epochs = 100)
```

```python
print(hist.history['loss'])
print(model.predict([[89., 91., 90.]]))
```
