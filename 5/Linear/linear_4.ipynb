```python
from keras.models import Sequential
from keras.layers import Dense
from keras import optimizers
import numpy as np
from sklearn.metrics import mean_squared_error    # 오차
```

```python
filename = '/content/drive/MyDrive/Colab Notebooks/Linear/score.csv'
data = np.loadtxt(filename, delimiter=',', dtype=np.float32)

X_data = data[ : , 0 : -1]   # slicing
Y_data = data[ : , [-1]]
```

```python
print(X_data, "\nX_data shape:", X_data.shape)  # x_data shape: (25, 3)
print(Y_data, "\nY_data shape:", Y_data.shape)  # y_data shape: (25, 1)
```

```python
train_X = data[0 : 20, 0:-1]
train_Y = data[0 : 20, [-1]]

test_X = data[20 : , 0:-1]
test_Y = data[20 : , [-1]]
```

```python
print(train_X, "\ntrain_X shape:", train_X.shape)  # train_X shape: (20, 3)
print(train_Y, "\ntrain_Y shape:", train_Y.shape)  # train_Y shape: (20, 1)

print(test_X, "\ntest_X shape:", test_X.shape)  # test_X shape: (5, 3)
print(test_Y, "\ntest_Y shape:", test_Y.shape)  # test_Y shape: (5, 1)
```

```python
# 모델 설계
model=Sequential()
model.add(Dense(1, input_dim = 3, activation = 'linear'))
model.summary()
```

```python
# 학습률 : 0.01 / sgd는 경사 하강법 / Cost function : 평균제곱오차 사용.
model.compile(optimizer = optimizers.SGD(lr = 0.000001), loss='mse')  # metrics = [‘accuracy’]

hist = model.fit(train_X, train_Y, batch_size =1, epochs = 100, validation_data=(test_X, test_Y))
```

```python
# 모델 평가 / 모델 검증
model.evaluate(test_X, test_Y)
print(model.predict([[78.,  83.,  77.]]))
predict = model.predict(test_X)
RMSE = mean_squared_error(test_Y, predict)**0.5
print("RMSE:", RMSE)
```

```python
import matplotlib.pyplot as plt
plt.plot(hist.history['loss'], 'r', label =  'loss')
plt.plot(hist.history['val_loss'], 'b', label =  'val_loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(loc='upper left')
plt.show()
```
