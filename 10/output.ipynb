```python
train_X = [[0.1, 4.2, 1.5, 1.1, 2.8],
           [1.0, 3.1, 2.5, 0.7, 1.1],
           [0.3, 2.1, 1.5, 2.1, 0.1],
           [2.2, 1.4, 0.5, 0.9, 1.1]]
print(np.shape(train_X))
```

```python
train_X=[[[0.1, 4.2, 1.5, 1.1, 2.8],
          [1.0, 3.1, 2.5, 0.7, 1.1],
          [0.3, 2.1, 1.5, 2.1, 0.1],
          [2.2, 1.4, 0.5, 0.9, 1.1]]]
train_X = np.array(train_X, dtype=np.float32)
print(np.shape(train_X))  # 1, 4, 5
```

```python
model = Sequential()
model.add(SimpleRNN(3, input_shape=(4, 5), return_state=False))
model.summary()

hidden_states=model(train_X)
print(hidden_states)
```

```python
model = Sequential()
model.add(SimpleRNN(3, input_shape=(4, 5), return_state=False))

hidden_states = model(train_X)
print(hidden_states)
```

```python
model = Sequential()
model.add(SimpleRNN(3, input_shape=(4, 5), return_sequences=True))
hidden_states = model(train_X)
print(hidden_states)
```

```python
model = Sequential()
model.add(LSTM(3, input_shape=(4, 5), return_state=False))

hidden_states = model(train_X)
print(hidden_states)
```

```python
model = Sequential()
model.add(LSTM(3, input_shape=(4, 5), return_sequences=True))
hidden_states = model(train_X)
print(hidden_states)
```
